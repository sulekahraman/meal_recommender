# Synthesizing Cuisine Specific Dataset

Create synthetic datasets that include certain number of recipes per cuisine.
Original dataset used: recipe_final.csv which much be unziped in the [data/original directory](../data/original)

```python
import pandas as pd
import numpy as np
```

```python
original_fn = '../data/original/recipe_final.csv'
data_final = pd.read_csv(original_fn)
data_final
```

```python
data_final[['ingredients']]
cuisines = set(data_final['cuisine'])
cuisines.remove(' garlic')
cuisines.remove('Cajun & Creole')
cuisines.remove(' tosted 125 mL 2 tbsp. chopped fresh cilantro 30 mL Directions 1\\xa0In tajine or large saucepan')
cuisines
```

```python
def make_data_for_each_cuisine(size=10):
    df = pd.DataFrame(columns=data_final.columns)
    for c in cuisines:
        #print(c)
        mask = (data_final['cuisine'] == c)
        #print(data_final[mask])
        df = df.append(data_final[mask][:size])
    df.to_csv('../data/synthetic/cuisine_size_{}.csv'.format(size))
```

```python
sizes = np.arange(10, 60, 10)
sizes
```

```python
for s in sizes:
    make_data_for_each_cuisine(size=s)
```
